```python
from scipy.optimize import curve_fit
from scipy.special import factorial

def poisson(k, lamb):
    return (lamb**k/factorial(k))*np.exp(-lamb)

degrees = [ER[0].degree(n) for n in ER[0].nodes()]
n, bins, patches= plt.hist(degrees, density=True, bins=18)

x=np.linspace(0,18, len(n))

parameters, cov_matrix = curve_fit(poisson, x, n)

print(parameters, cov_matrix)
plt.plot(x, poisson(x, *parameters), linewidth=2, label='$(\lambda^x/x!)e^{-\lambda}$')
plt.plot([], label=r'$\lambda=6.682 \pm 0.014$', color='w')
plt.xlabel('n', fontsize=15)
plt.ylabel('PDF', fontsize=15)
plt.legend(fontsize=13)
plt.show()
```
